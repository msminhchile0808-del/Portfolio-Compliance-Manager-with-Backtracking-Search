```python
"""
Portfolio Compliance Manager
Algorithm: Backtracking Search
"""

# --- 1. DATA ---
# Variables (X): The stocks we need to rate
stocks = ["AAPL", "MSFT", "XYL", "INTC","XOM", "CVX", "COP"]

# Domain (D): The values for each variable
domain = ["Buy", "Hold", "Sell"]

# Metadata used for constraints
sectors = { "AAPL": "Tech", "MSFT": "Tech", "XYL": "Tech", "INTC": "Tech",
           "XOM": "Energy", "CVX": "Energy", "COP": "Energy"}
is_esg_certified = {"AAPL": True, "MSFT": True, "XYL": True, "INTC": True,
                    "XOM": False, "CVX": False, "COP": False}

# --- 2. THE CONSTRAINT CHECK (The Rules) ---

def is_consistent(assignment):
    # Checks if the current partial assignment violates any rules (Constraints - Chapter 6).

    # Rule 1: Sector Risk Management (Max 2 'Buy' in Tech)
    tech_buys = [s for s, rating in assignment.items() if sectors[s] == "Tech" and rating == "Buy"]
    if len(tech_buys) > 2:
        return False

    # Rule 2: Diversification / Competitor Rule
    if assignment.get("AAPL") == "Buy" and assignment.get("MSFT") == "Buy" or assignment.get("XOM") == "Buy" and assignment.get("CVX") == "Buy":
        return False

    # Rule 3: ESG Mandate (Global Constraint)
    # If all stocks are assigned, we must have at least one ESG 'Buy'
    if len(assignment) == len(stocks):
        esg_buys = [s for s, rating in assignment.items() if is_esg_certified[s] and rating == "Buy"]
        if len(esg_buys) < 1:
            return False

    return True

# --- 3. BACKTRACKING SEARCH ALGORITHM ---

def backtrack(assignment):
    # Core CSP Solver: Recursively assigns values and backtracks on failure.

    # Goal Test: Are all stocks rated?
    if len(assignment) == len(stocks):
        return assignment

    # Selection: Pick the next unassigned stock
    var = [s for s in stocks if s not in assignment][0]

    for value in domain:
        assignment[var] = value

        # Only proceed if this rating doesn't break our rules
        if is_consistent(assignment):
            result = backtrack(assignment)
            if result:
                return result

        # Backtrack - Undo the choice if it leads to a dead end
        del assignment[var]

    return None

# --- 4. EXECUTION ---
print("🔍 Searching for a compliant portfolio allocation...")
solution = backtrack({})

if solution:
    print("✅ Found Compliant Portfolio:")
    for stock, rating in solution.items():
        print(f" - {stock}: {rating}")
else:
    print("❌ No compliant allocation found.")
```
